### total duration by season

```python
heist[['Season ','Watch Time (minutes)']].groupby('Season ').sum()['Watch Time (minutes)'].rename("Total watch time(minutes)")
```

- Grouping the season with sum
- Season-5 has the highest watch time (543)
- Low season-2 (279)
- And rename the **watch time** column to `Total watch time`

```python
df2=pd.DataFrame({"Total watchtime(minutes)":
heist[['Season ','Watch Time (minutes)']].groupby('Season ').sum()['Watch Time (minutes)'].rename("Total watchtime(minutes)").values},index=[1,2,3,4,5])
```

```python
sns.heatmap(heist[['Year','Watch Time (minutes)','IMDB']].corr(),annot=True,vmin=1)
plt.title('correlation between time,year,rating')
plt.show()
```

- It shows the correlation of columns Year,Watch Time (minutes),IMDB
- 1 means there is a `**absolute +ve**` relation
- +ve values have the +ve relation and -ve values have the -ve relation
